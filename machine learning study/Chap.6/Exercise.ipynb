## 問題6.1

```python
#ライブラリをインポート
import numpy as np
import pandas as pd
import statsmodels.api as sm
import math
```

ダミー変数を取り入れた表データはexercise.csvのようになる。  
x_a = 1 or 0, x_b = 1 or 0, x_c = 1 or 0としたのち  
x_aを消去したモデルを考えれば良い。  
以下、今まで通り解析を実行する。

```python
#データフレームの作成
df = pd.read_csv("exercise.csv", sep=",")
```

```python
df.head()
```

```python
#説明変数を取り出す
X = df.loc[:, ['xb','xc']].values

#ターゲット変数を取り出す
y = df['score'].values
```

```python
#モデルの作成、バイアス項を追加
model = sm.OLS(y, sm.add_constant(X))
results = model.fit()

#結果を表示
print(results.summary())
```

```python
#予測値を吐き出し
pred = results.predict(sm.add_constant(X))

print(pred)
```

F_0 = 5.0 > 2であるから、ダミー変数の集まり、つまり質的変数xはターゲット変数に効いていると判断できる。
